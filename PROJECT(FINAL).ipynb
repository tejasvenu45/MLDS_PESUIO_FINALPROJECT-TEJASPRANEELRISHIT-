```python
import numpy as np 
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('london_weather.csv')
```

```python
df
```

```python
data1=df.copy()
check_nan = data1.isnull().values.any()
check_nan
data2=data1.dropna()
data2=data2.drop_duplicates(subset="date")
```

```python
data2
```

```python
data3 = data2.drop_duplicates(['date'])
type=[]
for index, row in data3.iterrows():
    if row['mean_temp']>20 and row['precipitation']<1:
      type.append('sunny')
      
    elif row['cloud_cover']>5 and row['precipitation']>5:
      type.append('rainy')
    elif row['cloud_cover']>=7:
      type.append('cloudy')
    
    elif row['mean_temp']<5:
      type.append('cold')
    elif row['global_radiation']>15:
      type.append('high radiation')
    else:
      type.append('pleasant')
data2['type']=type
```

```python
data2
```

```python
import seaborn as sns
p = sns.PairGrid(data2,hue='type') # pairgrid allows you to draw a grid dof small subplots of diff types 
p.map_upper(sns.scatterplot) # the plots on the upper diagonal are scatterplots
p.map_lower(sns.kdeplot) #  the lower diagomal consist of density plots
p.map_diag(sns.histplot)  # the diagonal plots are histograms
p.add_legend()
```

```python
import seaborn as sns
data_copy = data2.copy()
del data_copy['type'] 
corr = data_copy.corr()  
sns.heatmap(corr, cmap='YlGnBu') 
```

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
data2.fillna(method ='ffill', inplace = True)

x = np.array(data2['global_radiation']).reshape(-1, 1)

y = np.array(data2['sunshine']).reshape(-1, 1)

data2.dropna(inplace = True)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

regr = LinearRegression()

regr.fit(x_train, y_train)

print(regr.score(x_test, y_test))

y_pred = regr.predict(x_test)

plt.scatter(x_test, y_test, color ='b')

plt.plot(x_test, y_pred, color ='k')

plt.show()
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print(r2)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
x=data1.iloc[:,:-1]
y=data1.iloc[:,-1]
from sklearn.model_selection import train_test_split
m_train, m_test, n_train, n_test = train_test_split(x, y, train_size=0.5)
classifier = KNeighborsClassifier(n_neighbors=7)
classifier.fit(m_train,n_train)
n_pred=np.round(classifier.predict(m_test))
```
